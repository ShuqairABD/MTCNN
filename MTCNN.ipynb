```python
# Импорт необходимых библиотек
import cv2
from matplotlib import pyplot as plt
from mtcnn import MTCNN
import numpy as np
from google.colab import files
```

**обнаружения лиц**

```python
# Функция для обнаружения лиц с использованием MTCNN
def detect_faces(image_path):
    # Загрузка изображения
    image = cv2.imread(image_path)
    # Конвертация изображения в RGB (требуется для MTCNN)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Инициализация детектора лиц MTCNN
    detector = MTCNN()

    # Обнаружение лиц на изображении
    faces = detector.detect_faces(image_rgb)

    # Нанесение прямоугольников вокруг лиц
    for face in faces:
        x, y, w, h = face['box']
        cv2.rectangle(image_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Конвертация изображения обратно в формат BGR для отображения с помощью OpenCV
    image_bgr = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)

    # Отображение изображения с нарисованными прямоугольниками
    plt.figure(figsize=(8, 8))
    plt.imshow(image_bgr)
    plt.axis('off')
    plt.show()

# Загрузка изображения в Colab
uploaded = files.upload()

# Получение пути загруженного изображения
image_path = next(iter(uploaded.keys()))

# Обнаружение лиц на загруженном изображении
detect_faces(image_path)
```
